```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# 生成 x 轴数据
x = np.linspace(0, 10, 100)

# 生成两条曲线的 y 轴数据
y1 = np.sin(x)
y2 = np.cos(x)

# 绘制两条曲线，设置不同颜色
plt.plot(x, y1, color='blue', label='sin(x)')
plt.plot(x, y2, color='red', label='cos(x)')
#plt.show()

# 线性回归预测
# 对 sin(x) 进行线性回归
x_reshaped = x.reshape(-1, 1)
model1 = LinearRegression()
model1.fit(x_reshaped, y1)
y1_pred = model1.predict(x_reshaped)
plt.plot(x, y1_pred, color='cyan', linestyle='--', label='sin(x) Linear Regression')

# 对 cos(x) 进行线性回归
model2 = LinearRegression()
model2.fit(x_reshaped, y2)
y2_pred = model2.predict(x_reshaped)
plt.plot(x, y2_pred, color='magenta', linestyle='--', label='cos(x) Linear Regression')

# 添加标题和标签
plt.title('Two Curves Plot with Linear Regression')
plt.xlabel('x')
plt.ylabel('y')

# 显示图例
plt.legend()

# 显示网格线
plt.grid(True)

# 显示图形
plt.show()
    
```
